```python
class Food:
    def __init__(self, name, protein, lemak, abu, serat, karbohidrat, air):
        self.name = name
        self.protein = protein
        self.lemak = lemak
        self.abu = abu
        self.serat = serat
        self.karbohidrat = karbohidrat
        self.air = air

def main():
    
    foods_hewani = [
        Food("Pakan 1", 10, 20, 5, 2, 15, 48),
        Food("Pakan 2", 15, 10, 8, 4, 18, 45)
    ]

    print("Daftar Pakan:")
    for i, food in enumerate(foods_hewani, start=1):
        print(f"{i}. {food.name}")

    choice = int(input("Pilih pakan yang anda gunakan: "))

    if choice > 0 and choice <= len(foods_hewani):
        selected_food = foods_hewani[choice - 1]
        print("\nInformasi Makanan:")
        print(f"Nama: {selected_food.name}")
        print(f"Protein: {selected_food.protein} %")
        print(f"Lemak: {selected_food.lemak} %")
        print(f"Abu: {selected_food.abu} %")
        print(f"Serat: {selected_food.serat} %")
        print(f"Karbohidrat: {selected_food.karbohidrat} %")
        print(f"Air: {selected_food.air} %")

        # Input komposisi pakan baru
        new_protein = float(input("Masukkan presentase protein baru: "))
        new_lemak = float(input("Masukkan presentase lemak baru: "))
        new_abu = float(input("Masukkan presentase abu baru: "))
        new_serat = float(input("Masukkan presentase serat baru: "))
        new_karbohidrat = float(input("Masukkan presentase karbohidrat baru: "))
        new_air = 100 - (new_protein + new_lemak + new_abu + new_serat + new_karbohidrat)

        # Menggabungkan komposisi pakan awal dengan pakan baru
        combined_food = Food("Pakan Gabungan", (selected_food.protein + new_protein / 2),
                             (selected_food.lemak + new_lemak / 2),
                             (selected_food.abu + new_abu / 2),
                             (selected_food.serat + new_serat / 2),
                             (selected_food.karbohidrat + new_karbohidrat / 2),
                             (selected_food.air + new_air / 2))

        print("\nInformasi Pakan Gabungan:")
        print(f"Nama: {combined_food.name}")
        print(f"Protein: {combined_food.protein} %")
        print(f"Lemak: {combined_food.lemak} %")
        print(f"Abu: {combined_food.abu} %")
        print(f"Serat: {combined_food.serat} %")
        print(f"Karbohidrat: {combined_food.karbohidrat} %")
        print(f"Air: {combined_food.air} %")

    else:
        print("Pakan tidak valid")

if __name__ == "__main__":
    main()
```
